<h3> Making plots using Altair </h3>

```python
brush = alt.selection_interval(encodings=['x','y'])
```

```python
chart1 = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"},
  "mark":"bar",
  "encoding":{
    "x":{"field":"City", "type":"nominal"},
    "y":{"aggregate":"sum","field":"Square Footage"},
    "color" : alt.value("#FFAA10")
  }
}).add_selection(
    brush
)
```

```python
chart2 = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"},
  "mark":"rect",
  "height":400,
  "encoding":{
    "x":{"bin":{"maxbins":10}, "field":"Square Footage", "type":"quantitative"},
    "y":{"field":"Agency Name", "type":"ordinal"}, 
    "color":{"aggregate":"sum", "type":"quantitative"} 
  } 
}).transform_filter(
    brush
)
```

```python
chart3 = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_bcubcg_fall2022/main/data/building_inventory.csv"},
  "mark":"line",
  "encoding":{
    "x":{"field":"Year Constructed", "type":"nominal"},
    "y":{"aggregate":"sum","field":"Square Footage"},
    "color" : alt.value('#7fc97f')
  }
}).transform_filter(
    brush
)
```

```python
chart = chart1 | chart2 | chart3
```

```python
chart
```

<h3> Plotting using ipywidgets </h3>

```python
@ipywidgets.interact(agency_name = data['Agency Name'].unique())
def create_scatter_plot(agency_name):
    streets_selected = data.loc[data['Agency Name'] == agency_name]
    streets_selected['city_zip'] = streets_selected['City'].astype('str') + ' ' + streets_selected['Zip code'].astype('str')
    
    sorted_sq = streets_selected.sort_values(by=['Square Footage'], ascending = False )
    
    plt.figure(figsize = (20,10))
    plt.subplot(3,1,1)
    plt.scatter(x ='Year Acquired', y = 'Year Constructed', data = streets_selected)
    plt.subplot(3,1,2)
    plt.bar(sorted_sq.head(20)['city_zip'], sorted_sq.head(20)['Square Footage'])
    plt.subplot(3,1,3)
    plt.bar(sorted_sq.head(20)['Year Constructed'],sorted_sq.head(20)['Square Footage'])
    plt.savefig(r"D:\IS455_Data_viz\Homework7\ketan96-m.github.io\assets\pngs\img.png")
    plt.show()
```

<p>The provided visualizations are created using ipywidgets and maplotlib. The first chart is a scatter plot and displays the correlation between the Year Contsructed and the Year Acquired for the selected Agency from the dropdown.</p>
<p>Chart 2 is a bar plot which displays the square footage for all the cities the selected Agency is in. The data has been tarnsformed to add a new column <i>City_zip</i> which contains the City and the different Zip Codes in order to show the square footage separately for each area. Chart 3 is a bar plot chart which displays the total square footage of the selected  agency against the year it was constructed.</p>

<p><b>Interactivity used:</b> Dropdown selection using <i>ipywidegts.interact</i> method.</p>
